## 1- Utilizando a rede do desafio:
# a- Itendifique comunidades usando o método Leiden ou Louvain.
Optamos por utilizar o método Louvain

```python
import community as community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import networkx as nx

#Exemplo base da biblioteca

# Rede do desafio
G = nx.read_gml("GraphMissingEdges.gml")

# usa o Louvain para computar a melhor partição
partition = community_louvain.best_partition(G)


# Plot do grafo

#Layout
pos = nx.spring_layout(G)


# Cores para representar uma comunidade em particular
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40, cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)

#nx.draw_networkx_labels(G, pos=pos)

plt.show()
```

# b- Identifique comunidades utilizando o k-clique. Utilize k=4.

```python
import networkx as nx
from networkx.algorithms.community import k_clique_communities

G = nx.read_gml("GraphMissingEdges.gml")

c = list(k_clique_communities(G, 4))

for i in c:
    print(sorted(list(i)))
```
