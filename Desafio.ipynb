```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gml('GraphMissingEdges.gml')

nx.draw(G, with_labels=True)
plt.show()
```

```python

clusters = nx.clustering(G)
print(f"O coeficiente de clusterizao geral: {nx.average_clustering(G)}")
```

```python
# distribuio de graus
degree_sequence = [d for n, d in G.degree()]

# histograma da distribuio de graus
degree_counts = nx.degree_histogram(G)

# distribuio de graus
plt.bar(range(len(degree_counts)), degree_counts, tick_label=range(len(degree_counts)))
plt.xlabel("Grau")
plt.ylabel("Nmero de Ns")
plt.title("Distribuio de Graus")
plt.show()
```
